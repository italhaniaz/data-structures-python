## <b> All and Any </b>

You can use <b> all </b> and <b> any </b> if you have a lot of conditions all or some of which need to be satisfied.

```python
a = 5
b = 4
```

```python
if a > 2 and a < 10 and b > 0 and b < 10:
    print("all constranints are satisfied")
```

```python
# any bussines logic
constraints = { a > 2, a < 10, b > 0, b < 10 }
```

```python
if all(constraints):
    print("All constraints are satisfied")
```

```python
if any(constraints):
    print("At least one constraint satisfied")
```

## <b> Counter </b>

A counter is the same as a dictinonary except us used to keep count and as such does not throw a keyError

```python
from collections import Counter
cnt = Counter()
for word in ['red', 'blue', 'red', 'green', 'blue', 'blue']:
    cnt[word] += 1
```

```python
print(cnt)
```

```python
cnt.most_common()
```

```python
cnt.most_common()[0][0]
```
