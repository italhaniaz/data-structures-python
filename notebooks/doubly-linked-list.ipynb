# <b> Doubly connected linked list </b>

A doubly connected linked list is quite similar to ta singly linked list. Just need to add a prev pointer to all the nodes.

```python
class Node:
    def __init__(self, data=None):
        self.val = data
        self.next = None
        self.prev = None
```

```python
class Doubly:
    def __init__(self):
        self.head = None
```

## Push Operation

```python
def push(self, value):
    new_node = Node(value)

    if self.head is None:
        self.head = new_node
        return

    #other wise do that
    temp = self.head
    while temp.next is not None:
        temp = temp.next

    temp.next = new_node
    new_node.prev = temp

Doubly.push = push
    
```

```python
# String operation

def __str__(self):
    ret_str = '['
    temp = self.head

    while temp is not None:
        # print (id(temp.prev), id(temp), id(temp.next))
        ret_str += str(temp.val) + ', '
        temp = temp.next

    ret_str = ret_str.rstrip(', ')
    ret_str += ']'

    return ret_str

Doubly.__str__ = __str__
```

```python
l = Doubly()
l.push(1)
l.push(2)
l.push(3)
print(l)
```

## Pop Operation

```python
def pop(self):
    if self.head is None:
        raise Exception("Cannot have any thing to pop")

    if self.head.next is None:
        ret_val = self.head.val
        self.head = None
        return ret_val

    #other cases
    last = self.head 
    while last.next is not None:
        prev = last
        last = last.next

    ret_val = last.val
    prev.next = None

    return ret_val

Doubly.pop = pop
```

```python
print(l)
l.pop()
l.pop()
print(l)
```

## Insert Operation

```python
def insert(self, index, value):
    new_node = Node(value)

    if index == 0:
        new_node.next = self.head
        if self.head is not None:
            self.head.prev = new_node

        self.head = new_node
        return 

    # for other cases
    temp = self.head
    index_coutner = 0
    while temp is not None and index_coutner < index:
        prev = temp
        temp = temp.next
        index_coutner += 1

    prev.next = new_node
    new_node.prev = prev

    new_node.next = temp
    if temp is not None:
        temp.prev = new_node

    return 

Doubly.insert = insert
```

### Assingment function -> REMOVE(val)

```python
def remove(self, val):

    if self.head is None:
        raise Exception("Cannot have any value to remove")

    if self.head.val == val:
        ret_val = self.head.val
        self.head = self.head.next

        if self.head is not None:
            self.head.prev = None

        return ret_val

    temp = self.head
    while temp is not None:
        if temp.val == val:
            break

        prev = temp
        temp = temp.next

    if temp is None:
        return

    ret_val = temp.val
    prev.next = temp.next
    if prev.next is not None:
        temp.next.prev = prev

    return ret_val

Doubly.remove = remove
```
