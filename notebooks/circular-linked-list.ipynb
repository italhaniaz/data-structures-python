# <b> Circular Linked list </b> 

we also called a circular linked list (or Ring data structure) 
This is also very similar to a regular linked list -- expect the last node is "linked" to the head thus creating a ring.

Due to this, we can't "loop unitl None is reached" because we will never have a none in the next now --even for a single node! so, we need to keep track of the head and always loop "until we reach back to the head"

```python
class Node:
    def __init__(self, data=None):
        self.val = data
        self.next = None
```

```python
class Ring:
    def __init__(self):
        self.head = None
```

```python
def __str__(self):
    ret_str = '['

    temp = self.head
    while temp is not None:
        ret_str += str(temp.val) + ', '
        temp = temp.next

        if temp == self.head:
            break

    ret_str = ret_str.rstrip(', ')
    ret_str += ']'

    return ret_str
```

```python
Ring.__str__ = __str__
```

```python
def _get_last(self):
    # if no node exist
    if self.head is None:
        return None

    # if one node exist
    if self.head.next == self.head:
        return self.head

    # at least two nodes
    temp = self.head.next
    while temp.next != self.head:
        temp = temp.next

    return temp
```

```python
Ring._get_last = _get_last
```

## Insert Operation

```python
def insert(self, index, val):
    new_node = Node(val)
    last = self._get_last()

    if index == 0:
        new_node.next = self.head
        self.head = new_node
        if last is None:
            self.head.next = self.head
        else:
            last.next = new_node

        return 

    if self.head is None and index > 0:
        raise IndexError("Cannot inset at" + str(index) + "because list is empty")

    # for other indices
    temp = self.head
    index_counter = 0
    while temp is not None and index_counter < index:
        prev = temp
        temp = temp.next
        index_counter += 1

    prev.next = new_node
    new_node.next = temp

    return 
```

```python
Ring.insert = insert
```

```python
r = Ring()
r.insert(0, 1)
r.insert(0, 2)
r.insert(1, 3)
r.insert(8, 5)
```

```python
print(r)
```

## Remove operation

```python
def remove(self, val):
    if self.head is None:
        raise Exception("Nothing to remove")

    temp = self.head
    last = self._get_last()
    # first node matches case
    if temp.val == val:
        if last == self.head:
            self.head = None
        else:
            self.head = temp.next
            last.next = self.head

        return

    prev = temp
    temp = temp.next
    while temp != last.next:
        if temp.val == val:
            break

        prev = temp 
        temp = temp.next

    if temp == last.next:
        return

    prev.next = temp.next

Ring.remove = remove
```

```python
print(r)
```

## Length operation

```python
def len(self):
    if self.head is None:
        return 0

    if self.head.next == self.head:
        return 1

    #for all other cases
    temp = self.head.next
    counter = 1
    while temp != self.head:
        counter += 1
        temp = temp.next

    return counter

Ring.len = len
```

```python
r.len()
```

```python
r.insert(5, -99)
r.insert(3, 4)
print(r)
r.len()
```

```python
r.remove(3)
r.remove(-99)
print(r)
r.len()
```

## Push operatoin

```python
def push(self, value):
    self.insert(self.len(), value)

Ring.push = push
```

```python
r.push(4)
print(r)
```

```python
r.push(9)
print(r)
```

## Pop operatoin

```python
def pop(self):
    if self.head is None:
        raise Exception("Nothing have to pop")

    if self.head.next == self.head:
        self.head = None
        return

    prev = self.head
    temp = self.head.next
    while temp.next != self.head:
        prev = temp 
        temp = temp.next

    prev.next = self.head

Ring.pop = pop
```

```python
print(r)
r.pop()
print(r)
```

## Get operation

```python
def get(self, index):
    if self.head == None:
        return 

    temp = self.head
    index_counter = 0
    while temp is not None and index_counter < index:
        temp = temp.next
        index_counter += 1
        
    return temp.val
```

```python
Ring.get = get
```
