```python
s = []

s.append(12) # <-- append is same as push
s.append(5)
s.append(55)

print(s)
print(s.pop())
print(s.pop())
print(s.pop())
# print(s.pop())  # <-- give index error
```

## Writing our own class of stack

```python
class Stack:
    def __init__(self):
        self.l = []
        
    def push(self, val):
        self.l.append(val)
        
    def pop(self):
        return self.l.pop()
        
    def peak(self):
        return self.l[-1]
```

## <b> Case Study : Bracket Matching </b>

```python
opening = '({['
closing = ')}]'
mapping = dict(zip(opening, closing))

print(mapping)
```

```python
print(mapping['('])
```

```python
def is_matched(string):
    opening = '({['
    closing = ')}]'
    mapping = dict(zip(opening, closing))

    stack = []
    
    for c in string:
        #case 1
        if c not in mapping.values() and c not in mapping.keys():
            continue

        # case 2
        # automatically checks the starting bracket

        if c in mapping:
            stack.append(mapping[c])

        # case 3: has to be closing bracket if we get here
        elif len(stack) == 0 or c != stack.pop():
            return False

    return len(stack) == 0
```

```python
string = "{[[]]{()}}"
is_matched(string)
```
